## 8. Detecting Outliers

```python
sns.boxplot(x = df['Playing'])
```

```python
sns.boxplot(x = df['Wishlist'])
```

```python
sns.boxplot(x = df['Rating'])
```

```python
df.describe()
```

```python
Q1 = df.Rating.quantile(0.25)
Q3 = df.Rating.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
lower_limit, upper_limit
```

```python
df[(df.Rating<lower_limit) | (df.Rating>upper_limit)]

#These are the outliers of Rating Column.
```

```python
#Removing the outliers From Rating column
df_no_outliers = df[(df.Rating>lower_limit) & (df.Rating<upper_limit)]
df_no_outliers
```

## 9. Plot different features against one another (scatter), against frequency (histogram)

```python
#Scatter Plot 
#to find the correlation between two variables
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['Plays'], df['Playing'])
ax.set_xlabel('Plays')
ax.set_ylabel('Playing')
plt.show()
```

```python
#Histogram 
df['Genres'].value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of Rating by Games Genres")
plt.ylabel('Rating')
plt.xlabel('Genres')
plt.show()
```
